```python
import glob
import pandas as pd
df = pd.DataFrame(columns=["content", "sentiment"])
for fn in glob.glob("data/aclImdb/train/pos/*.txt"):
    f = open(fn, "r", encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 1], index=["content", "sentiment"])
    df = df.append(s, ignore_index=True)
for fn in glob.glob("data/aclImdb/train/neg/*.txt"):
    f = open(fn, "r", encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 0], index=["content", "sentiment"])
    df = df.append(s, ignore_index=True)
df
```

```python
testdf = pd.DataFrame(columns=["content", "sentiment"])
for fn in glob.glob("data/aclImdb/test/pos/*.txt"):
    f = open(fn, "r", encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 1], index=["content", "sentiment"])
    testdf = testdf.append(s, ignore_index=True)
for fn in glob.glob("data/aclImdb/test/neg/*.txt"):
    f = open(fn, "r", encoding="utf-8")
    content = f.read()
    f.close()
    s = pd.Series([content, 0], index=["content", "sentiment"])
    testdf = testdf.append(s, ignore_index=True)
testdf
```

```python
from keras.preprocessing.text import Tokenizer
# 先創造你care的詞的詞典, 1000頁的辭典(等一下每個詞都是1000維度的向量)
tok = Tokenizer(num_words=1000)
tok.fit_on_texts(df["content"])
```

```python
# 到底1000個詞是哪一個1000個詞(抓出現次數最多的)
tok.word_index
```

```python
# 把每一篇文章的每一個詞都轉化成為數字(1000維度的向量)
train_seq = tok.texts_to_sequences(df["content"])
test_seq = tok.texts_to_sequences(testdf["content"])
# 因為每一篇文章不一樣多, 出現最多詞的文章1496個詞, 其他文章就會補上NaN
pd.DataFrame(train_seq)
```

```python
# 截長補短到每一篇文章都是同樣詞數
from keras.preprocessing.sequence import pad_sequences
train_pad = pad_sequences(train_seq, maxlen=100)
test_pad = pad_sequences(test_seq, maxlen=100)
# 如果你看到0, 補短所出現的結果
pd.DataFrame(train_pad)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Embedding, Dropout
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=32, input_length=100))
model.add(Flatten())
model.add(Dense(256, activation="relu"))
# Dropout 隨機丟棄一些資料
model.add(Dropout(0.25))
model.add(Dense(1, activation="sigmoid"))
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
model.fit(train_pad, df["sentiment"], 
          batch_size=200, epochs=3, 
          verbose=2, validation_split=0.1)
```

```python
model.evaluate(test_pad, testdf["sentiment"])
```

```python
# 儲存模型 .h5檔案
model.save("sentiment.h5")
```

```python
# 讀取模型(如果你要繼續訓練, 不用重新compile)
from keras.models import load_model
l = load_model("sentiment.h5")
l.evaluate(test_pad, testdf["sentiment"])
```

```python
# 你可以選擇中間層當作輸出
# model.layers[0] -> embedding層
# model.layers[0].output
# model.input 
from keras.models import Model
partial = Model(model.input, model.layers[0].output)
# [0] -> 取出第一篇文章 100個詞(32維度向量)
# [0][0] -> 取出第一篇文章 100個詞(32維度向量), 第一個詞
partial.predict(test_pad)[0][0]
```
